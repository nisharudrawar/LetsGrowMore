### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
df['variety']=label.fit_transform(df['variety'])
df.head()
```

### Data perparation

```python
X=df.drop(columns=['variety'])
y=df['variety']
```

### Train Test Split

```python
#split data in training and testing dataset
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
#printing sizes or training and testing dataset
print("X_train: ",X_train.shape)
print("X_test: ",X_test.shape)
print("y_train: ",y_train.shape)
print("y_test: ",y_test.shape)
```

### SVM Classification

```python
#importing svm
from sklearn.svm import SVC
svm = SVC()
```

```python
#predicting model using svm
model=svm
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#printing Accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ',accuracy)
```

```python
#classification report
cr = classification_report(y_test, y_pred)
print('Classification Report\n',cr)
```

### KNN Classification

```python
#importing knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
#Predition using knn model
model=knn
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
#printing Accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ',accuracy)
```

```python
#classification report
cr = classification_report(y_test, y_pred)
print('Classification Report\n',cr)
```

```python
print('Mean Absolute Error: ',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared Error: ',metrics.mean_squared_error(y_test, y_pred))
print('Root Mean squared Error: ', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#printing confusion matrix
matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(matrix,annot=True, fmt='', cmap='Blues')
plt.title('Confusion Matrix')
```

<B>CONCLUSION</B> We have applied svm and knn classification models on iris flowers dataset and got an accuracy of 98% from both the classifiers
