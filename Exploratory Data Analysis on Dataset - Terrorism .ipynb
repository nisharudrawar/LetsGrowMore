### Attacks Year wise

```python
#Countplot
plt.figure(figsize=(15,8))
sns.countplot(data = df, x = 'Year',palette='rocket')
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.xlabel("Year",size=20)
plt.ylabel("Count",size=20)
plt.show()
```

This plot counts Number of terrorist attacks per year<br>
According to the countplot Most number of terrorist attacks were seen happening in the year <B>2014</B> with attacks more than 16000
and least number of attacks were seen in the year <B>1973</B>.


### Attacks Region wise

```python
plt.subplots(figsize =(15,6))
sns.countplot('Region',data=df,palette='inferno',order=df['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Regions')
plt.title('Number of Terrorist Activities by Region')
plt.show()
```

As we can see the highest number of terrorist activities is in the Region of <B> South Asia and Middle,North Africa</B>

### Most Attacked Countries

```python
#Barplot
plt.subplots(figsize=(15,6))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='Blues_d')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.show()
```

This plot shows Top countries affected by terrorist attacks<br>
According to the above barplot <B>Iraq </B>is most affected by terrorist attacks. <B></B>

### Most Attacked Cities

```python
#Barplot
plt.figure(figsize=(15,9))
x =df['city'].value_counts().index[1:10]
y =df['city'].value_counts().values[1:10]
sns.barplot(x,y,palette='Pastel2')
plt.title('Top 10 Cities Affected By Terrorism',size=20)
plt.xlabel('Cities',size=15)
plt.ylabel('Count',size=15)
plt.xticks(size=13,rotation= 90)
plt.show()
```

This plot shows Top 10 cities affected by terrorist attacks<br>
<B>Conclusion</B> According to the above barplot <B>Baghdad</B> is the city most affected by terrorist attacks.

### Weapons Types

```python
plt.figure(figsize=(10,10))
df['Weapon_type'].value_counts().plot.pie(autopct ="%1.1f%%")
plt.title('Weapons used by the terrorist')
plt.show()
```

```python
#barplot
plt.figure(figsize =(15,7))
sns.barplot(x=df['Weapon_type'].value_counts()[:50].index,y =df['Weapon_type'].value_counts()[:50].values)
plt.xticks(rotation=90)
plt.xlabel('Weapons used', fontsize =22)
plt.ylabel("Frequency of a particular weapon used", fontsize =14)
plt.title("Type of Weapons", fontsize =22)
plt.show()
```

### Type of attacks

```python
plt.figure(figsize=(10,10))
df['AttackType'].value_counts().plot.pie(autopct ="%1.1f%%")
plt.title('Type of attacks used by terrorist')
plt.show()
```

```python
#barplot
plt.subplots(figsize =(15,6))
sns.countplot('AttackType',data=df,palette='inferno',order=df['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel('Attack Type')
plt.title('Attacking Methods by Terrorists')
plt.show()
```

### Most Target Types

```python
Target = df['Target_type'].value_counts().nlargest(n=15)
Target
```

```python
#Barplot
plt.figure(figsize= (10,8))
sns.barplot(x = Target.index , y = Target.values,palette='PuBuGn' )
plt.title('Most Frequent Target Type')
plt.xlabel('Target Type')
plt.ylabel('Attacks count')
plt.xticks(rotation = 90)
plt.show()
```

### Terrorist Activities by Region in each Year

```python
#crosstab
pd.crosstab(df.Year, df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```
